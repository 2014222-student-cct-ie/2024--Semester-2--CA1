```python
# Utilise Python programming language in order to comply with the requisites of
# the assessment and perform adequate Machine
# Learning algorithms to discover and deliver insights.

# Import the necessary libraries (Numpy and Pandas) in order to perform data cleansing.
# These are the libraries that are conventionally used as a common practice in order to
# perform mathematical and statistical operations during a data analysis process

import numpy as np
import pandas as pd


# Import requests to apply Data scraping, which is also called "web scraping,"
# basically I am geeting the information from a website
# It's one of the fastest ways to get information from the web

import requests

# I am using this line of code to see all columns in a wide DataFrame

pd.set_option('display.max_columns',None)

# I am using time to measures how long will take to  scraping the data of the NBA seasons
# I will use a for loop

import time


# Import the itertools library is because it simplifies the code by replacing nested loops
# with a single loop that iterates over the Cartesian product,
# making the code easier to read and understand.

from itertools import product
```

# Offical NBA website with the statistics.

```python
# I am using the official NBA website to pull my data in order to complete this CA1.
# The first step was to find which data in the NBA website I was going to use
# and then I had to inspect elements in the browser (Google Chrome)
# and go to the tab Network to find the right data and the specific location
# within an API endpoint that accepts requests and sends back responses

# Reading and testing the NBA url

nba_api_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2012-13&SeasonType=Regular%20Season&StatCategory=PTS'
```

# Using API and json

```python
# Request to scrap date from the NBA url using a json format
# So then I can manipulate to pull into a DataFrame

request = requests.get(url=nba_api_url).json()
```

```python
# Creating a table list with resutls and headers to user as columns in when creating a DataFrame format

table_headers = request['resultSet']['headers']
```

```python
# Creating a DataFrame format

pd.DataFrame (request['resultSet']['rowSet'], columns = table_headers)
```

```python
# Creating the foundation of DataFrame1
# Creating a new table that has the same number of rows as foundation_dataframe1
# Doing this is going to help me to add consistent information related to resutls and the row set

foundation_dataframe1 = pd.DataFrame (request['resultSet']['rowSet'], columns = table_headers)
```

```python
# Creating a new table that has the same number of rows as foundation_dataframe1
# Doing this is going to help me to add consistent information related to a specific year and season

# I am using a for loop on the new dataframe with repetitive data,
# making sure it matches the size (in terms of rows) of another dataframe. 

foundation_dataframe2 = pd.DataFrame ({'Year':['2012-13' for each_position_in_the_list in range(len (foundation_dataframe1))],
                                       'Season_type':['Regular%20Season' for each_position_in_the_list in range(len (foundation_dataframe1))]})
```

```python
# Print table with columns year and season type

foundation_dataframe2
```

```python
# Creating a new table called foundation_dataframe3.
# Which has all the columns from foundation_dataframe2 (like 'Year' and 'Season_type')
# followed by all the columns from foundation_dataframe1.

# I am using pd.concat() to sticks tables together.
# I wanted to attach them side by side.

foundation_dataframe3 = pd.concat([foundation_dataframe2, foundation_dataframe1], axis = 1)
```

```python
dataframe_columns_short = request['resultSet']['headers']
```

# Using for loop

```python
# Defining dataframe_columns.

dataframe_columns = ['Year','Season_type'] + dataframe_columns_short

# This will be the foundation of the main DataFrame

main_dataframe = pd.DataFrame(columns = dataframe_columns)

# Lists to iterate through

season_types = ['Regular%20Season', 'Playoffs']

years = ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22']

# Creating the for loop with itertools.product for combinations of years and season_types
# The reason why I am using the itertools lubrary is because it simplifies the code by replacing
# nested loops with a single loop that iterates over the Cartesian product,
# making the code easier to read and understand.

# Also, it makes it easy to add more dimensions.
# For example, if I want to scrape data for more variables (e.g., adding teams to season_types and years),
# I can simply add the new list to the product call without changing the loop structure.

start_loop = time.time()

for y, s in product(years, season_types):
    
    # Defining the NBA url API
    nba_api_url = f'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season={y}&SeasonType={s}&StatCategory=PTS'
    
    # Pulling out the json
    request = requests.get(url=nba_api_url).json()
    
    foundation_dataframe1 = pd.DataFrame(request['resultSet']['rowSet'], columns=dataframe_columns_short)
    
    foundation_dataframe2 = pd.DataFrame({'Year': [y for _ in range(len(foundation_dataframe1))],
                                           'Season_type': [s for _ in range(len(foundation_dataframe1))]})
    
    foundation_dataframe3 = pd.concat([foundation_dataframe2, foundation_dataframe1], axis=1)
    
    main_dataframe = pd.concat([main_dataframe, foundation_dataframe3], axis=0)
    
    print(f'Finished scraping data from the {y} {s}.')
    
    # To prevent the NBA website from blocking access for scraping data, adding a lag time
    # I am adding a lag time with the time package so the NBA website don't think I am a bot
    
    lag = np.random.uniform(low=20, high=40)
    print(f'...waiting {round(lag,1)} seconds...')
    time.sleep(lag)

# Print final message

print(f'Congratulations! Process completed! Total run time: {round(time.time() - start_loop,1)} seconds')
```
